##### Descripción General

-Una aplicación de un sistema no linear de ecuacion diferenciales en un ambiente matemático/biológico seria el modelo de la relación predador-presa de un ecosistema. En clase vimos la clásica ecuación propuesta por Lotka-Volterra 
 
$$\frac{dx}{dt} = a\; x\;- b\;x\;y\; $$

$$\frac{dy}{dt} = c\; x\;- d\;x\;y\;$$

Una desventaja de este modelo es que la población de la presa crecería sin restricciones en la ausencia de predador ( y\; = 0). Esto es un modelo idealista pero no realista. A continuación expondremos 3 variantes a este problema. 


Cabe destacar que trataremos a *$x$* como la especie presa y a *$y$*  cómo la especie predador en todos los casos


##### Objetivo General 

- Determinar el comportamiento de diferentes especies en determinadas circunstancias:

    -En el caso de que la presa sea puramente herbívoro y el depredador sobreviva solamente por la existencia de la presa.
    
    -En el caso de que la presa sea puramente herbívoro y el depredador omnívoro, es decir, su alimentacion no dependa solamente de la presa.
    
    -En el caso de un ecosistema cerrado con 3 especies, las cuales se comportan como una cadena. 


##### Objetivos Específicos

- Obtención de ecuaciones 
- Mostrar la gráfica de los 3 modelos, herbívoro-presa, herbívoro-omnívoro y cadena alimenticia
- Determinar como se comportan las especies en el tiempo, dadas las condiciones iniciales y las características del ecosistema y animales.
- Marcar las restricciones, es decir, que limitantes tienen los modelos.

## Modelo matemático  y simulación

### Caso 1

En nuestra primera variacion del modelo queremos eliminar el comportamiento exponencial de la presa en la ausencia del predador, por lo cual remplazamos el crecimiento exponencial de la primera ecuacion por una ecuación logística de crecimiento de 2 términos:


$$\frac{dx}{dt} = (a\; x\;- r\; x\;* x\;)- b\;x\;y\; $$

$$\frac{dy}{dt} = -c\; x\; + d\;x\;y\;$$

Donde **r** es la capacidad de carga del ecosistema.

En este caso, en la ausencia de depredadores, la primera ecuación se convierte en la ecuación logística, es decir, la población de la presa se estabilizaría al llegar a la capacidad de carga del ecosistema. 



En nuestro caso modelaremos la siguiente ecuación:

$$\frac{dx}{dt} = (1\; x\;- 5\; x\;*x\;)- 1\;x\;y\; $$

$$\frac{dy}{dt} = -0.5\; x\; + 0.75\;x\;y\;$$
 
 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t,a1,b1,c1,a2,b2,c2):
    x1 = x[0]
    x2 = x[1]
    return [a1*x1-b1*x1*x1-c1*x1*x2, -a2*x2+c2*x1*x2]

x0 = [1,1]

t = np.linspace(0, 10,300)

#parametros
a1=1
b1=5
c1=1
a2=1/2
c2=3/4

x = odeint(modelo, x0, t, args=(a1,b1,c1,a2,b2,c2))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('Cantidad de especies')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=3)
plt.grid()
plt.xlabel('$x_1$  Especie x')
plt.ylabel('$x_2$ Especie y')

plt.show()
```

## Caso 2

En nuestra segunda variacion del modelo queremos eliminar el comportamiento exponencial de la presa en la ausencia del predador, por lo cual remplazamos el crecimiento exponencial de la primera ecuación por una ecuación logística de crecimiento de 2 términos.
En este caso, en la ausencia de depredadores, al igual que en el primer caso se convierte en la ecuación logística, es decir, la población de la presa se estabilizaría al llegar a la capacidad de carga del ecosistema y la población de el depredador, al ser este un animal omnívoro, es decir, su alimento no depende solamente en la existencia de la presa tambien llegaria a la capacidad de carga en ausencia de la presa pero en abundancia de su otro alimento. 

Donde **r** es la capacidad de carga del ecosistema


$$\frac{dx}{dt} = (a\; x\;- r\; x\;* x\;)- b\;x\;y\; $$

$$\frac{dy}{dt} = (a\; y\;- r\; y\;* y\;) + d\;x\;y\;$$



En nuestro caso modelaremos la siguiente ecuación 
$$\frac{dx}{dt} = (1\; x\;- 5\; x\;*x\;)- 1\;x\;y\; $$

$$\frac{dy}{dt} = (0.5\; y\;- 0.25\; y\;* y\;) + 0.75\;x\;y\;$$
 
 



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t,a1,b1,c1,a2,b2,c2):
    x1 = x[0]
    x2 = x[1]
    return [(a1*x1-b1*x1*x1)-c1*x1*x2, (a2*x2-b2*x2*x2)+c2*x1*x2]

x0 = [1,1]

t = np.linspace(0,10,100)

#parametros
a1=1
b1=5
c1=1
a2=1/2
b2=1/4
c2=3/4

x = odeint(modelo, x0, t, args=(a1,b1,c1,a2,b2,c2))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutos]')
plt.ylabel('Comportamieto de especies en el tiempo')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=3)
plt.grid()
plt.xlabel('$x_1$  Especie x')
plt.ylabel('$x_2$ Especie y')
plt.show()
```

## Caso 3

En este caso modelaremos un ecosistema cerrado con 3 especies. La especie **x** se alimenta solamente de pasto (herbíviro) cuya población obedecería la ecuacion logística en caso de ausencia de depredadores. La especie **y** se alimenta de la especie **x**, sin embargo, esta es a su vez cazada por la especie **z**.
Podemos ver que **x** y **z**nunca interactuan directamente. 

$$\frac{dx}{dt} = (a\; x\;- b\; x\;* x\;)- c\;x\;y\; $$

$$\frac{dy}{dt} = -d\; y\;+ e\; y\;x\;- f\; y\;z\; $$

$$\frac{dz}{dt} = - g\; z\;+ h\; y\;z\; $$


En nuestro caso modelaremos la siguiente ecuación:

$$\frac{dx}{dt} = (3\; x\;- 3\; x\;* x\;)- 3\;x\;y\; $$

$$\frac{dy}{dt} = -1\; y\;+ 1.2\; y\;x\;- 1.4\; y\;z\; $$

$$\frac{dz}{dt} = - 2\; z\;+ 10\; y\;z\; $$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t, a, b, c, d, e, f, g, h):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    return [(a*x1-b*x1*x1)-c*x1*x2, -d*x2+e*x1*x2-f*x2*x3, -g*x3+h*x2*x3]

x0 = [1,1,0.5]

t = np.linspace(0, 10, 300)

#parametros
a = 3
b = 3
c = 3
d = 1
e = 1.2
f = 1.4
g = 2
h= 10


x = odeint(modelo, x0, t, args=(a,b,c,d,e,f,g,h))
x1 = x[:,0]
x2 = x[:,1]
x3 = x[:,2]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.plot(t, x3 , 'g', lw=2, label='$x_2(t)$: Especie z')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutos]')
plt.ylabel('Comportamieto de especies en el tiempo')
plt.show()
```

```python
# Ejemplo gráfico 3d
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt



# Esto no importa, es la generación de los datos x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 300)
z = np.linspace(-2, 2, 300)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

# Esta parte es la que permite graficar z vs (x,y)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x1, x2, x3, color='g', label='comportamiento')
ax.set_xlabel(' Especie x')
ax.set_ylabel(' Especie y')
ax.set_zlabel(' Especie z')
ax.legend()

plt.show()
```

# Conclusiones 


Podemos concluir que:

- En el caso 1 en donde controlamos el comportamiento exponencial de la especie **x** y con las condiciones iniciales que utilizamos que en este caso fueron  $\left[1,1\right]$. 
- Como limitante pudimos observar que la especie **x** crece desmedidamente en la ausencia de **y**, cosa que no sucede en la vida real. 
- En t= 5 existe un punto donde hay la misma cantidad de especies, sin embargo pasando este punto la presa comienza a ser mayor llegado a un equlibrio en **0.2**, llevando al depredador casi a la extinción. 



- En el caso 2 en donde la especie **x** se comporta con el control de la ecuación logística de dos términos anterior y la especie **y** es omnívora, es decir, en caso de que la especie **x** sea **0** aun puede existir y con las condiciones iniciales que utilizamos que en este caso fueron  $\left[1,1\right]$. 
- Pudimos observar que dadas las condiciones iniciales y las condiciones base, el predador acabaría con la presa y esta podría sobrevivir a base de su otro alimento, llegando asi a la capacidad del ecosistema. 
- Como limitante podemos observar que la ecuación de ambos depende del la suposición que exista abundancia de recursos para ambos, siendo ese el caso. Tanto presa como depredador se limitarían a comer solamente este alimento. 


- En el caso 3 donde apreciamos una cadena alimenticia, es decir la especie **x** es puramente herbívora, la especie **y** es omnívora y la especie **z** se alimenta exclusivamente de la especie **y** y con las condiciones iniciales que utilizamos que en este caso fueron  $\left[1,1,1\right]$. y las condiciones variables. 
- Pudimos observar que la especie **x** primero sufre una gran caida, pero en cuanto existe una sobrepoblación de depredador **z** esta lleva a la extincion a **y** provocando que **x** llegue a su capacidad de carga en el ecosistema.
