##### Descripción General

-Una aplicacion de un sistema no linear de ecuacion diferenciales en un abiente matemático/biológico seria el modelo de la relacion predador-presa de un ecosistema. Clase vimos la clasica ecuacion propuesta por Lotka-Volterra 
 
$$\frac{dx}{dt} = a\; x\;- b\;x\;y\; $$

$$\frac{dy}{dt} = c\; x\;- d\;x\;y\;$$

Una desventaja de este modelo es que la poblacion de la presa creceria sin restricciones en la ausencia de predador ( y\; = 0). Esto es un modelo ilealista pero no realista. A continuacion expondremos 3 variantes a este problema. 


Cabe destacar que trataremos a *$x$* como la especie presa y a *$y$*  como la especie predador en todos los casos


##### Objetivo General 

- Determinar el comportamiento de diferentes especies en determinadas circunstancias:

    -En el caso de que la presa sea puramente hervivora y el depredador sobreviva solamente por la existencia de la presa.
    
    -En el caso de que la presa sea puramente hervivora y el depredador omnívoro, es decir, su alimentacion no dependa solamente de la presa.
    
    -En el caso de un ecosistema cerrado con 3 especies, las cuales se comportan como una cadena. 


##### Objetivos Específicos

- Obtencion de ecuaciones 
- Mostrar la grafica de los 3 modelos, hervivoro-presa, hervívoro-omnívoro y cadena alimenticia
- Determinar como se comportan las especies en el tiempo, dadas las condiciones iniciales y las caracteristicas del ecosistema y animales.
- Marcar las restricciones, es decir, que limitantes tienen los modelos.

## Modelo matemático  y simulación

### Caso 1

En nuestra primera variacion del modelo queremos eliminar el comportamient exponencial de la presa en la ausencia del predador, por lo cual remplazamos el crecimiento expoencial de la primera ecuacion por una ecuacion logistica de crecimiento de 2 términos:


$$\frac{dx}{dt} = (a\; x\;- r\; x\;* x\;)- b\;x\;y\; $$

$$\frac{dy}{dt} = -c\; x\; + d\;x\;y\;$$

Donde **r** es la capacidad de carga del ecosistema

En este caso, en la absencia de depredadores, la primera ecuacion se convierte en la ecuacion logistica, es decir, la poblacion de la presa se estabilizaría al llegar a la capacidad de carga del ecosistema. 



En nuestro caso modelaremos la siguiente ecuacion:

$$\frac{dx}{dt} = (1\; x\;- 5\; x\;*x\;)- 1\;x\;y\; $$

$$\frac{dy}{dt} = -0.5\; x\; + 0.75\;x\;y\;$$
 
 


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t,a1,b1,c1,a2,b2,c2):
    x1 = x[0]
    x2 = x[1]
    return [a1*x1-b1*x1*x1-c1*x1*x2, -a2*x2+c2*x1*x2]

x0 = [1,1]

t = np.linspace(0, 10,300)

#parametros
a1=1
b1=5
c1=1
a2=1/2
c2=3/4

x = odeint(modelo, x0, t, args=(a1,b1,c1,a2,b2,c2))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t')
plt.ylabel('Cantidad de especies')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=3)
plt.grid()
plt.xlabel('$x_1$  Especie x')
plt.ylabel('$x_2$ Especie y')

plt.show()
```

## Caso 2

En nuestra segunda variacion del modelo queremos eliminar el comportamient exponencial de la presa en la ausencia del predador, por lo cual remplazamos el crecimiento expoencial de la primera ecuacion por una ecuacion logistica de crecimiento de 2 términos.
En este caso, en la ausencia de depredadores, al igual que en el primer caso se convierte en la ecuacion logistica, es decir, la poblacion de la presa se estabilizaría al llegar a la capacidad de carga del ecosistema y la poblacion de el depredador, al ser este un animal omnivoro, es decir, su alimento no depende solamente en la existencia de la presa tambien llegaria a la capacidad de carga en ausencia de la presa pero en abundancia de su otro alimento. 

Donde **r** es la capacidad de carga del ecosistema


$$\frac{dx}{dt} = (a\; x\;- r\; x\;* x\;)- b\;x\;y\; $$

$$\frac{dy}{dt} = (a\; y\;- r\; y\;* y\;) + d\;x\;y\;$$



En nuestro caso modelaremos la siguiente ecuacion 
$$\frac{dx}{dt} = (1\; x\;- 5\; x\;*x\;)- 1\;x\;y\; $$

$$\frac{dy}{dt} = (0.5\; y\;- 0.25\; y\;* y\;) + 0.75\;x\;y\;$$
 
 



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t,a1,b1,c1,a2,b2,c2):
    x1 = x[0]
    x2 = x[1]
    return [(a1*x1-b1*x1*x1)-c1*x1*x2, (a2*x2-b2*x2*x2)+c2*x1*x2]

x0 = [1,1]

t = np.linspace(0,10,100)

#parametros
a1=1
b1=5
c1=1
a2=1/2
b2=1/4
c2=3/4

x = odeint(modelo, x0, t, args=(a1,b1,c1,a2,b2,c2))
x1 = x[:,0]
x2 = x[:,1]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutos]')
plt.ylabel('Comportamieto de especies en el tiempo')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(x1, x2, lw=3)
plt.grid()
plt.xlabel('$x_1$  Especie x')
plt.ylabel('$x_2$ Especie y')
plt.show()
```

## Caso 3

En este caso modelaremos un ecosistema cerrado con 3 especies. La especie **x** se alimenta solamente de pasto (hervíviro) cuya poblacion obedeceria la ecuacion logistica en caso de ausencia de depredadores. La especie **y** se alimenta de la especie **x**, sin embargo, esta es a su vez cazada por la especie **z**.
Podemos ver que **x** y **z**nunca interactuan directamente. 

$$\frac{dx}{dt} = (a\; x\;- b\; x\;* x\;)- c\;x\;y\; $$

$$\frac{dy}{dt} = -d\; y\;+ e\; y\;x\;- f\; y\;z\; $$

$$\frac{dz}{dt} = - g\; z\;+ h\; y\;z\; $$


En nuestro caso modelaremos la siguiente ecuacion:

$$\frac{dx}{dt} = (a\; x\;- b\; x\;* x\;)- c\;x\;y\; $$

$$\frac{dy}{dt} = -d\; y\;+ e\; y\;x\;- f\; y\;z\; $$

$$\frac{dz}{dt} = - g\; z\;+ h\; y\;z\; $$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def modelo(x, t, a, b, c, d, e, f, g, h):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    return [(a*x1-b*x1*x1)-c*x1*x2, -d*x2+e*x1*x2-f*x2*x3, -g*x3+h*x2*x3]

x0 = [1,1,0.5]

t = np.linspace(0, 10, 300)

#parametros
a = 3
b = 3
c = 3
d = 1
e = 1.2
f = 1.4
g = 2
h= 10


x = odeint(modelo, x0, t, args=(a,b,c,d,e,f,g,h))
x1 = x[:,0]
x2 = x[:,1]
x3 = x[:,2]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, x1 , 'r', lw=2, label='$x_1(t)$: Especie x')
plt.plot(t, x2 , 'b', lw=2, label='$x_2(t)$: Especie y')
plt.plot(t, x3 , 'g', lw=2, label='$x_2(t)$: Especie z')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutos]')
plt.ylabel('Comportamieto de especies en el tiempo')
plt.show()
```
